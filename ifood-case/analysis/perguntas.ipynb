##### Qual a média de valor total (total\_amount) recebido em um mês considerando todos os yellow táxis da frota?



O primeiro ponto é filtrar apenas pelas corridas da categoria yellow.
Temos que agregar por mês, somando o total_amount das corridas. 
Com isso, temos o total de cada mês, então, tiramos a média desses valores

```python
%sql
with combinado_mensal AS (
  select
    source_month, 
    sum(total_amount) AS soma
  from
    silver_unified_trips
  where 
    trip_category = 'yellow'
  group by
    source_month
)
select
  avg(soma)
FROM
  combinado_mensal;
```

##### Qual a média de passageiros (passenger_count) por cada hora do dia que pegaram táxi no mês de maio considerando todos os táxis da frota?

Basta filtrar pelo mês correspondente e agrupar pela hora de pickup, pegando a média dos passageiros.
No fim, ordenei pela hora por legibilidade.

```python
%sql
select 
    DATEPART('hour', tpep_pickup_datetime) as hora, 
    avg(passenger_count)
from 
    silver_unified_trips
where
    source_month = '2023-05'
group by 
    hora
order by 
    hora asc
```
